## Data extraction (web scraping), data organization

```python
import pandas as pd
import numpy as np
from urllib.request import urlopen
import json
```

```python

url_first='https://pubchem.ncbi.nlm.nih.gov/sdq/sdqagent.cgi?infmt=json&outfmt=json&query={%22select%22:%22*%22,%22collection%22:%22compound%22,%22where%22:{%22ands%22:[{%22*%22:%22EGFR%22}]},%22order%22:[%22relevancescore,desc%22],%22start%22:1,%22limit%22:10,%22width%22:1000000,%22listids%22:0}'
response_first = urlopen(url_first)
data_json0 = json.loads(response_first.read())
page=data_json0['SDQOutputSet'][0]['rows']
page_count=int(data_json0['SDQOutputSet'][0]['totalCount']/10)


cid_list=[]
cmpdname_list=[]
mf_list=[]
isosmiles_list=[]
for elem in page:
    cid_list.append(elem['cid'])
    cmpdname_list.append(elem['cmpdname'])
    mf_list.append(elem['mf'])
    isosmiles_list.append(elem['isosmiles'])


for p in range(1,page_count+1):  
    #I used double {{}} to escape the syntax error
    url_rest=f'https://pubchem.ncbi.nlm.nih.gov/sdq/sdqagent.cgi?infmt=json&outfmt=json&query={{%22select%22:%22*%22,%22collection%22:%22compound%22,%22where%22:{{%22ands%22:[{{%22*%22:%22EGFR%22}}]}},%22order%22:[%22relevancescore,desc%22],%22start%22:{p}1,%22limit%22:10,%22width%22:1000000,%22listids%22:0}}'
    response_rest=urlopen(url_rest)
    data_json=json.loads(response_rest.read())
    page_output=data_json['SDQOutputSet'][0]['rows']
    
    for elem in page_output:
        cid_list.append(elem['cid'])
        cmpdname_list.append(elem['cmpdname'])
        mf_list.append(elem['mf'])
        isosmiles_list.append(elem['isosmiles'])
    

cid_list
```

```python
##Create a dataframe in python and save it to csv file
dic={'cid':cid_list,'cmpdname':cmpdname_list,'mf':mf_list,'isosmiles':isosmiles_list}
df=pd.DataFrame(dic)
df
```
