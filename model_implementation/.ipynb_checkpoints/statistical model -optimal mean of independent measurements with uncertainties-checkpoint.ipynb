## implementation
Randomly generate data

```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
##Generate random data set with n (self-defined) measurements
n=1000
repeats=10
X=[]
for i in range(repeats):
    x=random.sample(range(0,1000),n)
    X.append(x)
X
```

```python
##variance of each measurement [delta^2 xi]:
v=[]
for measurements in X:
    variance=0
    for point in measurements:
        variance+=(point-np.mean(measurements))**2
    v.append(variance)
```

```python
##with the derived formular for optimal variance:
```

```python
##sum of reciprocal
sum(np.reciprocal(v))
```

```python
## the optimal variance of Independent Measurements with Uncertainties equals to:
1/sum(np.reciprocal(v))
```
